<h1 style="color:black">7-6 7-9  Pivot basics</h1>

```python
import pandas as pd
import numpy as np
df = pd.read_csv("7-9_weather.csv")
df.head(2)
```

```python
df.pivot(index='city',columns='date')
```

```python
df.pivot(index='city',columns='date',values="humidity")
```

```python
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='humidity',columns='city')
```

<h1 style="color:black">Pivot Table</h1>

```python
df = pd.read_csv("7-9_weather2.csv")
df.head(2)
```

```python
df.pivot_table(index="city",columns="date")
```

<h2 style="color:black">Margins</h2>

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
```

<h2 style="color:black">Grouper</h2>

```python
df = pd.read_csv("7-9_weather3.csv")
df.head(2)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
